```python
from bs4 import BeautifulSoup as bs
from splinter import Browser
import os
import pandas as pd
import time
import pymongo
import requests
```

```python
# READING IN 20 ZILLOW PAGES FOR ALL SANTA CLARA COUNTY PROPERTIES ABOVE $1.4 MILLION
# PUT THOSE LISTINGS IN THE LIST Santa_Clara_Houses

Santa_Clara_Houses = []
```

```python
executable_path = {'executable_path': '/usr/local/bin/chromedriver'}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
url = 'https://www.zillow.com/santa-clara-county-ca/luxury-homes/21_p/'
browser.visit(url)
```

```python
html = browser.html
soup = bs(html,'html.parser')
```

```python
list_cards = soup.find_all('div', class_='list-card-info')
```

```python
for listing in list_cards:
    try:
        table = []
        address = listing.find('address', class_="list-card-addr").text
        price = listing.find('div', class_="list-card-price").text
        beds = listing.find('ul', class_="list-card-details")
        lis = beds.find_all('li')
        table.append(lis)
        beds = lis[0].text
        baths = lis[1].text
        square_feet = lis[2].text
        house_dict = {"address": address, "price": price, "beds": beds, "baths": baths, "square feet": square_feet}
        Santa_Clara_Houses.append(house_dict)
                
        if (price):
            print('-------------')
            print(address)
            print(price)
            print(beds)
            print(baths)
            print(square_feet)
    
    except IndexError as e:
        print(e)
```
